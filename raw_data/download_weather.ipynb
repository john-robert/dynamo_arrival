```python
import pandas as pd
import json
import urllib
import urllib.request
```

```python
station_dresden = "01048"
```

```python
first_date = "2019-01-01"
```

```python
last_date = "2020-12-31"
```

```python
url = f"https://api.brightsky.dev/weather?date={first_date}&last_date={last_date}&dwd_station_id={station_dresden}"
req = urllib.request.Request(url)
response = urllib.request.urlopen(req)
data = response.read()
values = json.loads(data) 
```

```python
df_new= pd.DataFrame(values['weather'])
```
